```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('/Users/shreyaravichandiran/Documents/GitHub/Train-Complaint-System/grievances.csv')
```

```python
# Display the first few rows of the dataset
print(df.head())
```

```python
# Remove duplicate rows (if any)
df = df.drop_duplicates()
```

```python
# Drop rows with missing values
df = df.dropna()
```

```python
df[' Incident date'] = pd.to_datetime(df[' Incident date'], format='%YYYY-%mm-%dd', errors='coerce')  # Adjust format if needed
df[' PNR no'] = df[' PNR no'].astype(str)
df['Mobile no'] = df['Mobile no'].astype(str)
```

```python
# Save the cleaned dataset to a new CSV file
df.to_csv('grievances_cleaned.csv', index=False)
```

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('/Users/shreyaravichandiran/Documents/GitHub/Train-Complaint-System/grievances_cleaned.csv')

# Define the priority mapping
priority_mapping = {
    'Medical Assistance': 1,
    'Security': 1,
    'Divyangjan Facilities': 2,
    'Facilities for Women with Special needs': 2,
    'Electrical Equipment': 3,
    'Coach - Cleanliness': 3,
    'Punctuality': 4,
    'Water Availability': 4,
    'Coach - Maintenance': 5,
    'Catering & Vending Services': 5,
    'Staff Behaviour': 6,
    'Corruption / Bribery': 7,
    'Bed Roll': 8,
    'Miscellaneous': 9
}

# Replace missing values with 'Miscellaneous'
df[' Type'] = df[' Type'].fillna('Miscellaneous')

# Map priority
df['Priority'] = df[' Type'].map(priority_mapping)

# Handle missing values in the 'Priority' column
df.loc[:, 'Priority'] = df['Priority'].fillna(df['Priority'].max() + 1)

# Sort the dataframe by priority
df_sorted = df.sort_values(by='Priority')

# Print all data in the CSV file in an orderly manner based on priority
print(df_sorted)
```
